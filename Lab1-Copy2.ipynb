```python
poke.columns = ['no','name', 'type1', 'type2', 'total', 'hp', 'attack', 'defense','spatk','spdef','speed','gen','legend']
```

```python

x=poke.total
y=poke.hp

plt.plot(x,y,'o',color='green')
```

```python
def plot(x, y, ax, title, y_label):
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.plot(x,y)
    ax.margins(x=0,y=0)
plt.plot(x,y,'o',color='black')
```

```python
#to count how many unique types there are
poke['type1'].nunique()
poke['gen'].nunique()

#make a list of the unique types
t1_name = poke.type1.unique()
print(t1_name)
```

```python
#how to count unique items using Counter
from collections import Counter
Ctype1=poke['type1']
Counter(Ctype1)
```

```python
#How to make a Pie Chart using .csv unique values

from collections import Counter

poke['type1'].value_counts().plot.pie()

plt.title("Frequencies of Pokemon Types")
plt.startangle=90 #Not quite sure what this does
```

```python
#Same data as above, but as a bar plot

from collections import Counter

poke['type1'].value_counts().plot.bar()
```
